```python
#importar as bibliotecas
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
import csv
```

```python
#link copiado com o botao direito, para copiar o link do download
#Define a url
url = "https://download.inep.gov.br/microdados/microdados_enem_2019.zip"

#Faz o download do conteudo e a faz a requisicao pro download
filebytes = BytesIO(
    requests.get(url).content
    ) 

#extrair o conteúdo do zipfile
myzip = zipfile.ZipFile(filebytes)
myzip.extractall("./enem2019") #extrai tudo nessa pasta
```
